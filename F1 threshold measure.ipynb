```python
import pandas as pd
from scipy import stats

# Calculate correlation between drs_active and stability_index
# Point-biserial correlation is used for one binary and one continuous variable
correlation, p_value = stats.pointbiserialr(df_sorted_asc['drs_active'], df_sorted_asc['stability_index'])

print(f"Correlation Coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4e}")

# Grouped mean for context
grouped_means = df_sorted_asc.groupby('drs_active')['stability_index'].mean()
print("\nAverage Stability Index by DRS Status:")
print(grouped_means)
```

```python
def calculate_stability_threshold(df):
    """
    let's take the maximum downforce number that can be recorded without braking the stability
    of the car.
    """
    # Identify the first row where stability is no longer perfect (100)
    instability_point = df[df['stability_index'] < 100].head(1)

    if instability_point.empty:
        return "No instability detected in the current dataset.", None

    threshold_n = instability_point['downforce_n'].values[0]
    speed_at_limit = instability_point['speed_kmh'].values[0]
    wing_at_limit = instability_point['wing_angle_deg'].values[0]

    return threshold_n, {
        'speed': speed_at_limit,
        'wing_angle': wing_at_limit,
        'stability': instability_point['stability_index'].values[0]
    }

# Execute calculation
limit_n, details = calculate_stability_threshold(df_sorted_asc)

if details:
    print("--- Downforce Stability Threshold Platform ---")
    print(f"Maximum Safe Downforce Limit: {limit_n:.2f} N")
    print(f"Critical Speed reached: {details['speed']:.2f} km/h")
    print(f"Critical Wing Angle: {details['wing_angle']:.2f} deg")
    print(f"Initial Stability Drop: {details['stability']:.2f}%")

    # Apply flagging to the dataset for further analysis
    df_sorted_asc['stability_status'] = df_sorted_asc['downforce_n'].apply(
        lambda x: 'SAFE' if x < limit_n else 'STABILITY_WARNING'
    )
else:
    print(limit_n)
```
