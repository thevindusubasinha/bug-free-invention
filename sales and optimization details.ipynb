```python
import pandas as pd

# Load the dataset
df_fmcg = pd.read_csv('/content/fmcg_sales_3years_1M_rows.csv')

# Filter rows where stock_out_flag is 1
stock_out_df = df_fmcg[df_fmcg['stock_out_flag'] == 1]

# Display the first few rows of the filtered table
display(stock_out_df.head())

# Show the shape to confirm count
print(f'Total rows with stock_out_flag == 1: {len(stock_out_df)}')
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# 1. Filter for events where margin_pct is negative to calculate losses
loss_df = df_fmcg[df_fmcg['margin_pct'] < 0].copy()
# Calculate the absolute dollar value of the lost profit
loss_df['lost_profit_value'] = loss_df['net_sales'] * loss_df['margin_pct'].abs()

# 2. Aggregate profit and loss by brand
brand_profit = df_fmcg.groupby('brand')['profit'].sum().reset_index()
brand_loss = loss_df.groupby('brand')['lost_profit_value'].sum().reset_index(name='total_loss')

# 3. Create a summary table for comparison
profit_loss_summary = pd.merge(brand_profit, brand_loss, on='brand', how='left').fillna(0)
profit_loss_summary = profit_loss_summary.rename(columns={'profit': 'total_profit'})

# 4. Generate side-by-side bar charts
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Total Profit by Brand', 'Total Lost Profit by Brand (Margin < 0)'))

# Green bars for overall profit
fig.add_trace(
    go.Bar(name='Total Profit', x=profit_loss_summary['brand'], y=profit_loss_summary['total_profit'],
           marker_color='green', text=profit_loss_summary['total_profit'], texttemplate='$%{y:,.2f}', textposition='outside'),
    row=1, col=1
)

# Red bars for calculated losses
fig.add_trace(
    go.Bar(name='Total Lost Profit', x=profit_loss_summary['brand'], y=profit_loss_summary['total_loss'],
           marker_color='red', text=profit_loss_summary['total_loss'], texttemplate='$%{y:,.2f}', textposition='outside'),
    row=1, col=2
)

fig.update_layout(
    title_text='Brand Financial Performance: Profit vs. Losses (3 Years)',
    height=500,
    showlegend=False,
    hovermode='x unified'
)

fig.update_yaxes(title_text='Amount ($)', row=1, col=1)
fig.update_yaxes(title_text='Amount ($)', row=1, col=2)

fig.show()
```

This interactive bar chart visually compares the total profit and total lost profit for each brand. This side-by-side comparison helps in understanding the overall financial health and pricing effectiveness of each brand.

```python
import pandas as pd

# Ensure date is in datetime format
df_fmcg['date'] = pd.to_datetime(df_fmcg['date'])

# Extract day of week (0=Monday, 6=Sunday)
df_fmcg['day_of_week'] = df_fmcg['date'].dt.dayofweek

# Define Weekday vs Weekend (5=Saturday, 6=Sunday)
df_fmcg['period_type'] = df_fmcg['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Group by Year and Period Type to see variations
weekly_variation = df_fmcg.groupby(['year', 'period_type']).agg(
    total_profit=('profit', 'sum'),
    total_units_sold=('units_sold', 'sum'),
    avg_margin_pct=('margin_pct', 'mean'),
    transaction_count=('date', 'count')
).reset_index()

# Calculate Year-over-Year changes for profit
weekly_variation['profit_formatted'] = weekly_variation['total_profit'].map('${:,.2f}'.format)

print('Financial Variation: Weekdays vs Weekends (2021-2023)')
display(weekly_variation)
```

```python
import pandas as pd

# Group by year and period_type to compare behavior
# We normalize units sold by the number of days (5 for Weekday, 2 for Weekend)
behavior_comparison = df_fmcg.groupby(['year', 'period_type']).agg(
    avg_units_per_period=('units_sold', 'sum'),
    avg_margin_pct=('margin_pct', 'mean'),
    avg_discount_pct=('discount_pct', 'mean'),
    total_days=('date', 'nunique')
).reset_index()

# Calculate daily averages to make the comparison fair
behavior_comparison['avg_daily_units_sold'] = behavior_comparison['avg_units_per_period'] / behavior_comparison['total_days']

# Format for display
behavior_comparison_display = behavior_comparison[[
    'year', 'period_type', 'avg_daily_units_sold', 'avg_margin_pct', 'avg_discount_pct'
]].copy()

behavior_comparison_display['avg_margin_pct'] = (behavior_comparison_display['avg_margin_pct'] * 100).map('{:.2f}%'.format)
behavior_comparison_display['avg_discount_pct'] = (behavior_comparison_display['avg_discount_pct'] * 100).map('{:.2f}%'.format)
behavior_comparison_display['avg_daily_units_sold'] = behavior_comparison_display['avg_daily_units_sold'].map('{:,.0f}'.format)

print('Buying Behavior Comparison: Daily Averages (2021-2023)')
display(behavior_comparison_display)
```

```python
import plotly.express as px
import pandas as pd

# 1. Group by year, brand, subcategory, and period_type to get totals and day counts
brand_year_avg = df_fmcg.groupby(['year', 'brand', 'subcategory', 'period_type']).agg(
    total_units=('units_sold', 'sum'),
    days_count=('date', 'nunique')
).reset_index()

# 2. Calculate daily average to normalize the comparison
brand_year_avg['avg_daily_units'] = brand_year_avg['total_units'] / brand_year_avg['days_count']

# 3. Create a combined label for the x-axis
brand_year_avg['Product_Label'] = brand_year_avg['brand'] + ' (' + brand_year_avg['subcategory'] + ')'

# 4. Filter for the top 15 products by average volume to keep the chart readable across years
top_products = brand_year_avg.groupby('Product_Label')['avg_daily_units'].mean().sort_values(ascending=False).head(15).index
filtered_avg_data = brand_year_avg[brand_year_avg['Product_Label'].isin(top_products)]

# 5. Create the interactive facet plot (separated by year)
fig = px.bar(
    filtered_avg_data,
    x='Product_Label',
    y='avg_daily_units',
    color='period_type',
    facet_row='year',
    barmode='group',
    title='Average Daily Units Sold by Year: Weekday vs. Weekend',
    labels={'avg_daily_units': 'Avg Units Per Day', 'Product_Label': 'Product'},
    height=900,
    color_discrete_map={'Weekday': '#636EFA', 'Weekend': '#EF553B'}
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
